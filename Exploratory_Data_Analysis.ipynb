Statistik Deskriptif

```python
display(df.describe())
```

Data Visualization

```python
# Buat histogram untuk kolom numerik
numerical_cols = ['artist_pop', 'track_pop', 'danceability', 'energy', 'loudness']

df[numerical_cols].hist(bins=15, figsize=(15, 10))
plt.suptitle('Histograms of Numerical Features', y=1.02)
plt.tight_layout()
plt.show()
```

Data Cleaning

```python
#Check Missing Value
display(df.isnull().sum())
```

```python
#Check Duplicate Word
display(df.duplicated().sum())
```

```python
# Check unique values for categorical columns to identify inconsistencies
categorical_cols = ['artist_name', 'album']

for col in categorical_cols:
    print(f"Unique values in '{col}':")
    display(df[col].unique()[:50]) # Displaying up to 50 unique values
    print("-" * 30)

# Note: For columns with many unique values, displaying all might not be practical.
# You might need to further inspect specific values if inconsistencies are suspected.
```

```python
# Check data scale for numerical columns
numerical_cols = df.select_dtypes(include=['number']).columns

print("Data Scale (Min, Max, Mean, Std Dev) for Numerical Columns:")
display(df[numerical_cols].agg(['min', 'max', 'mean', 'std']))
```

```python
# Calculate skewness for numerical columns
numerical_cols = df.select_dtypes(include=['number']).columns

print("Skewness of Numerical Columns:")
display(df[numerical_cols].skew())
```

```python
# Buat box plot terpisah untuk setiap kolom numerik agar lebih mudah dibaca skalanya
numerical_cols = df.select_dtypes(include=['number']).columns

for col in numerical_cols:
    plt.figure(figsize=(8, 4)) # Sesuaikan ukuran figure untuk setiap plot
    sns.boxplot(x=df[col])
    plt.title(f'Box Plot of {col}')
    plt.xlabel(col)
    plt.show()
```

Korelasi Dataset

```python
# Hitung matriks korelasi untuk kolom numerik
correlation_matrix = df.select_dtypes(include=['number']).corr()

# Buat heatmap korelasi
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

Berdasarkan langkah-langkah eksplorasi data yang telah dilakukan, berikut adalah rangkuman temuan utama dari dataset TikTok songs 2022:

1. Ukuran Dataset
Dataset terdiri dari 263 baris (lagu) dan 18 kolom, mencakup berbagai metadata dan fitur audio dari lagu-lagu TikTok.

2. Data Cleaning: Dataset ini relatif bersih. Tidak ditemukan adanya missing value (nilai yang hilang) maupun baris duplikat. Pemeriksaan nilai unik pada kolom kategorikal artist_name dan album juga tidak menunjukkan inkonsistensi yang signifikan dalam format atau penulisan pada sampel yang dilihat.

3. Statistik Deskriptif dan Distribusi:
- Fitur seperti danceability, energy, dan loudness memiliki rentang nilai yang bervariasi, dengan rata-rata danceability sekitar 0.71, menunjukkan bahwa lagu-lagu TikTok cenderung mudah diiringi tarian.
- Populeritas artis (artist_pop) dan lagu (track_pop) juga bervariasi, dengan beberapa lagu dan artis memiliki popularitas yang sangat tinggi (mendekati 100), sementara ada juga yang populeritasnya rendah.
- Kolom speechiness, acousticness, instrumentalness, dan liveness menunjukkan skewness positif yang signifikan, yang berarti sebagian besar lagu memiliki nilai rendah pada fitur-fitur ini, dengan beberapa outlier bernilai tinggi. instrumentalness memiliki skewness tertinggi.
- time_signature mayoritas adalah 4, dengan sedikit kasus 1, 3, dan 5.

4. Visualisasi Distribusi (Histogram & Box Plot):
- Histogram menunjukkan distribusi dari beberapa fitur numerik. Misalnya, distribusi danceability cenderung condong ke kiri (skewed negatif ringan), sementara energy dan loudness distribusinya lebih mendekati normal atau sedikit miring ke kiri.

- Box plot secara jelas menunjukkan sebaran setiap kolom numerik dan mengidentifikasi adanya outlier pada banyak kolom, termasuk artist_pop, track_pop, danceability, energy, loudness, speechiness, acousticness, instrumentalness, dan liveness. Outlier ini mungkin mewakili lagu atau artis yang karakteristiknya sangat berbeda dari mayoritas data.

- Hubungan Antar Variabel (Pair Plot & Heatmap Korelasi)
1. Pair plot memberikan gambaran visual tentang hubungan antara pasangan fitur numerik yang dipilih.
2. Heatmap korelasi menunjukkan kekuatan dan arah hubungan linier antar kolom numerik. Beberapa korelasi yang patut dicatat antara lain:
3. Korelasi positif yang cukup kuat antara energy dan loudness (0.59). Ini wajar karena lagu dengan energi tinggi seringkali juga lebih keras.
4. Korelasi positif moderat antara artist_pop dan track_pop (0.53), menunjukkan bahwa popularitas artis berkorelasi dengan popularitas lagu mereka di TikTok.
5. Korelasi negatif moderat antara acousticness dengan energy (-0.46) dan loudness (-0.39). Lagu akustik cenderung memiliki energi dan kenyaringan yang lebih rendah.
6. danceability memiliki korelasi positif yang lemah hingga moderat dengan valence (0.37), menunjukkan bahwa lagu yang lebih mudah diiringi tarian cenderung memiliki suasana yang lebih positif.
